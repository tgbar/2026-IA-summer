# Notebook 2 — Cosmological distances

In this notebook we compute cosmological distances and see how they depend on the dark energy model.

For a flat universe the key distances are:

$$d_c(z) = \frac{c}{H_0}\int_0^z \frac{dz'}{E(z')} \qquad \text{(comoving distance)}$$

$$d_L(z) = (1+z)\, d_c(z) \qquad \text{(luminosity distance)}$$

The function `solve_distances()` computes both in a single pass.

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import *
from plots import *
```

---
## A minimal example

Compute and plot $d_L(z)$ for ΛCDM.

```python
Om = 0.3
rho_de, _ = de_model('LCDM', Om)
z, d_c, d_L = solve_distances(Om, rho_de)

print(f"d_L(z=1) = {np.interp(1.0, z, d_L):.4f} c/H0  =  {np.interp(1.0, z, d_L)*DH():.0f} Mpc")
plot_dL(z, d_L, Om)
```

## Comparing models

`utils.py` provides three dark energy models:
- `'LCDM'` — cosmological constant ($w = -1$)
- `'CPL'` — Chevallier-Polarski-Linder, parameters `w0` and `wa`
- `'lCDM'` — $\lambda$CDM quintessence (Fonseca et al. 2021), parameter `lam`

The luminosity distance is sensitive to the dark energy model — this is how we can distinguish them observationally.

```python
Om  = 0.3
w0  = -0.9
wa  =  0.2
lam =  0.5

models = [
    ('LCDM', {},                   'ΛCDM'),
    ('CPL',  {'w0': w0, 'wa': wa}, f'CPL ($w_0={w0}, w_a={wa}$)'),
    ('lCDM', {'lam': lam},         f'$\\lambda$CDM ($\\lambda={lam}$)'),
]

fig, ax = plt.subplots(figsize=(9, 5))
for model, kwargs, label in models:
    rho, _ = de_model(model, Om, **kwargs)
    z, d_c, d_L = solve_distances(Om, rho)
    ax.plot(z, d_L * DH(), label=label)

ax.set_xlabel(r'redshift $z$')
ax.set_ylabel(r'$d_L(z)$  [Mpc]')
ax.set_title(f'$\\Omega_m={Om}$, flat, $h=0.7$')
ax.legend()
plt.tight_layout()
plt.show()
```

# Notebook 1 — The expanding universe

In this notebook we explore the dynamics of the universe using the Friedmann equation.
All the numerical machinery lives in `utils.py`, plotting routines in `plots.py`.

We assume a **flat universe** by default ($\Omega_k = 0$), so $\Omega_\Lambda = 1 - \Omega_m$.

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import *
from plots import *
```

---
## A minimal example

Standard ΛCDM, flat universe, $\Omega_m = 0.3$.

```python
Om = 0.3
rho_de, drho_de = de_model('LCDM', Om)
t, a, adot = solve_friedmann(Om, rho_de, drho_de)

print(f"Age of the universe: {-t[0]:.4f} H0^-1  ({-t[0]*tH():.2f} Gyr)")
plot_at(t, a, Om)
```

---
## Exploring further

`plots.py` also provides:

- `plot_phase(a, adot, Om)` — phase portrait $\dot{a}(a)$
- `plot_adotdot(t, a, Om, rho_de, drho_de)` — acceleration $\ddot{a}(t)$
- `plot_densities(a, Om, rho_de)` — energy densities vs $N = \ln a$

```python
plot_densities(a, Om, rho_de)
```

```python
plot_phase(a, adot, Om)
```

```python
plot_adotdot(t, a, Om, rho_de, drho_de)
```

```python
plot_adot(t, adot, Om)
```

---
## Adding curvature

Pass `Ok` to include curvature. $\Omega_k > 0$ is open, $\Omega_k < 0$ is closed.

```python
Om = 0.3
Ok = -0.1   # closed universe
rho_de, drho_de = de_model('LCDM', Om, Ok=Ok)
t, a, adot = solve_friedmann(Om, rho_de, drho_de, Ok=Ok)

print(f"Age of the universe: {-t[0]:.4f} H0^-1  ({-t[0]*tH():.2f} Gyr)")
plot_at(t, a, Om, Ok)
```

## A different dark energy model

`utils.py` provides three dark energy models via `de_model()`:
- `'LCDM'` — cosmological constant ($w = -1$)
- `'CPL'` — Chevallier-Polarski-Linder, parameters `w0` and `wa`
- `'lCDM'` — $\lambda$CDM quintessence (Fonseca et al. 2021), parameter `lam`

```python
Om = 0.3
rho_de, drho_de = de_model('CPL', Om, w0=-0.9, wa=0.2)
t, a, adot = solve_friedmann(Om, rho_de, drho_de)

print(f"Age of the universe: {-t[0]:.4f} H0^-1  ({-t[0]*tH():.2f} Gyr)")
plot_at(t, a, Om)
```
